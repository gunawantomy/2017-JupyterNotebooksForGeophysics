```python
# Preparation for programming
# Make sure to execute this cell first!
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')                  # do not show warnings
from IPython.display import display, Math, Latex
from obspy import read
from obspy.core import read, UTCDateTime
from obspy.clients.fdsn import Client
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
from datetime import datetime 
```

```python
start=datetime.now()

sta_list=pd.read_csv('stationlist.csv', sep=',',header=None)
sta_code=sta_list[0]
sta_chan=sta_list[1]
sta_tra=sta_list[2]
sta_net=sta_list[3]

client = Client("IRIS")
t = UTCDateTime("2004-12-26T00:58:53.0")#origin time of Sumatra earthquake
starttime1 =t # origin time of Sumatra earthquke
endtime1 = t+100 # 600s after the the origin time
for i in range(len(sta_code)):
    sta_code[i] = client.get_waveforms(sta_net[i], sta_code[i], sta_tra[i], sta_chan[i], starttime1, endtime1, attach_response=True)
    sta_code[i].write('File%s.sac' %i, format='SAC')
    print(sta_code[i])
    
d = datetime.now()-start
print (d, " s")
```
